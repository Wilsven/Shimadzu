```python
import pandas as pd
```

```python
# Read xlsx file
df = pd.read_excel('Data Gathering.xlsx')
df
```

```python
# Drop useless columns
df = df.drop(['ID', 'Start time', 'Completion time', 'Email', 'Name'], axis = 1)
df
```

```python
# Rename Name2 to Name and sub-cat to 1-4 respectively
df = df.rename(columns = {'Name2': 'Name',
                         'Sub-Category2': '2',
                         'Sub-Category3': '3',
                         'Sub-Category4': '4',
                         'Customer': 'Type'})
df
```

```python
# Replace NaN values with '-'
df = df.fillna('-')
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid")

sns.countplot(x='Category', data=df)
plt.show()
```

```python
df.loc[:,'2':'4']
```

```python
target = df.loc[:,'2':'4']
d = {}
for index in target.index:
    for i in target.iloc[index]:
        if i == '-':
            continue
        else:
            d[index] = i

d
```

```python
df['Sub-Category'].tolist()
```

```python
copy = df['Sub-Category'].tolist().copy()
for i, v in enumerate(copy):
    if i in d:
        copy[i] = d[i]
    else:
        continue
        
# Replace
df['Sub-Category'] = copy
```

```python
df
```

```python
# Remove redundant 2, 3, 4 columns
df = df.drop(['2', '3', '4'], axis = 1)
df
```

```python
# Repeat for customer and project columns
# Combine project with customer columns into new column

target_proj = df['Project']
d_proj = {}
for i in target_proj.index:
    if target_proj[i] == '-':
        continue
    else:
        d_proj[i] = target_proj[i]
        
d_proj # only has three projects B, E and D
```

```python
copy_type = df['Type'].tolist().copy()
for i, v in d_proj.items():
    copy_type[i] = v
    
copy_type
```

```python
# Replace
df['Type'] = copy_type
df
```

```python
# Remove Project column
df = df.drop(['Project'], axis=1)
df
```
